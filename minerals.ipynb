```python
from bs4 import BeautifulSoup as bs
import pandas as pd
from urllib import urlopen
import numpy
from pprint import pprint
```

```python
minerals_url = "http://minerals.usgs.gov/minerals/pubs/historical-statistics/"
minerals_html = urlopen(minerals_url).read()
```

```python
soup = bs(minerals_html, "lxml")
```

```python
xlsx_links = soup.find_all('a', text='XLSX')
```

```python
xlsxs = {} #braces are for dictionaries

i = 0
for filename in xlsx_links:
    base_url = "http://minerals.usgs.gov/minerals/pubs/historical-statistics/" + filename['href']
    base_url_obj = urlopen(base_url)
    base_xls = pandas.read_excel(base_url_obj)
    xlsxs[filename['href']] = base_xls
    print filename['href'], i
    i += 1
```

```python
mineral_production_by_year = {}
for df in xlsxs.keys():
    header = list(xlsxs[df].irow(3).values)
    i = 0
    world_production_by_year = {}
    for h in header:
        try:
            if 'World production' in h:
                world_production_by_year['World production'] = list(xlsxs[df].icol(i).values)[4:]
            if 'Year' in h: 
                world_production_by_year['Year'] = list(xlsxs[df].icol(i).values)[4:]
            i+=1
            mineral_production_by_year[df] = world_production_by_year
        except:
            i+=1
            continue
    
pprint(mineral_production_by_year)
```
